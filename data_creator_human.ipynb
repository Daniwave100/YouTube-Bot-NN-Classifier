```python
# ...existing code...
import random

# choose a different sample slice each run to vary examples
start = random.randint(0, max(0, len(df) - 30))
sample_data = '\n'.join(df.text.iloc[start:start + 15].astype(str))

# use non-zero temperature to encourage variety
temperature = 0.2

content_prompt = f"""
Analyze the contents of the following DataFrame.
The DataFrame contains real scraped comments from YouTube financial or money-related videos.
Each comment here is a real human comment.
You are creating original human comment conversation threads to be used as training data for a machine learning model.
Only output the conversation messages, one message per line — do NOT add any leading text like "Sure, here's a conversation thread" or any separators.
Vary the style, length and number of messages between threads. Do not repeat the exact same phrasing across outputs.
Talk about different financial topics, ask questions, and provide opinions.
Do not only talk about stocks or cryptocurrencies, but also about general financial advice, market trends, and personal finance like real estate and 401(k) plans.

Here are example messages to mimic:
{sample_data}

IMPORTANT:
- Output only the conversation thread messages, one per line.
- Vary the number of messages and tone across outputs.
- Do NOT include any extra commentary or separators.
Example:
CORRECT:
Charlie, what are your thoughts on VISA stock right now? I think it's a good buy given the current market.
"""
```

```python
model = "gpt-4o-mini"
temperature = 0

# number of conversation threads to generate
num_threads = 500

csv_frame = []
```

```python
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI()
thread_id_num = 100

for i in range (num_threads):
    print(f"Generating thread {i + 1} of {num_threads}")
    
    # Create a response using the OpenAI client
    response = client.responses.create(
        model=model,
        temperature=temperature,
        input=content_prompt
    )

    llm_output = response.output[0].content[0].text
    thread = llm_output.split("\n")

    for comment in thread:
        csv_frame.append({
            "prompt": content_prompt,
            "temperature": temperature,
            "model": model,
            "thread number": thread_id_num,
            "text": comment,
            "author": "Human",
            "boolean": 1,
            })

    thread_id_num += 1

csv_pd_frame = pd.DataFrame(csv_frame)
csv_final = csv_pd_frame.to_csv("generated_data_human.csv", index=False)
```

```python
data = pd.read_csv("generated_data_human.csv")
data.head()
```

```python
rows_to_drop = []

for index, row in data.iterrows():  # iterate over rows in dataframe:
    if pd.isna(row["text"]):
        rows_to_drop.append(index)
    elif "---" in str(row["text"]):
        rows_to_drop.append(index)
    elif "Sure!" in str(row["text"]):
        rows_to_drop.append(index)
    elif "Feel free to ask" in str(row["text"]):
        rows_to_drop.append(index)

data.drop(rows_to_drop, inplace=True)
```
