```python
import pandas as pd

bot_data = pd.read_csv("generated_data.csv")
human_data = pd.read_csv("generated_data_human.csv")
```

```python
human_data.head()
```

```python
print(len(bot_data))
print(len(human_data))
```

```python
# Get 20% sample to create test set
bot_test = bot_data.sample(frac=0.2, random_state=42)
human_test = human_data.sample(frac=0.2, random_state=42)

# Get remaining 80% of data for our training set
bot_train = bot_data.drop(bot_test.index)
human_train = human_data.drop(human_test.index)

# Concatenate our training and test sets and shuffle them
train_data = pd.concat([bot_train, human_train], ignore_index=True).sample(frac=1, random_state=42)
test_data = pd.concat([bot_test, human_test], ignore_index=True).sample(frac=1, random_state=42)
```

```python
# create new CSVs just to have them
train_data.to_csv("train_data.csv", index=False)
test_data.to_csv("test_data.csv", index=False)
```
